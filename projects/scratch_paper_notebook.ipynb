```python
import matplotlib.pyplot as plt

# Data: Global EV sales numbers from 2010 to 2023
years = list(range(2010, 2024))
sales = [
    17000,   # 2010
    55000,   # 2011
    130000,  # 2012
    220000,  # 2013
    315000,  # 2014
    550000,  # 2015
    750000,  # 2016
    1200000, # 2017
    2000000, # 2018
    2200000, # 2019
    3100000, # 2020
    6600000, # 2021
    10000000,# 2022
    14000000 # 2023
]

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(years, sales, marker='o', linestyle='-', color='b')
plt.title('Global EV Sales from 2010 to 2023')
plt.xlabel('Year')
plt.ylabel('EV Sales (in units)')
plt.grid(True)
plt.xticks(years, rotation=45)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for TAM, SAM, and SOM projections
years = list(range(2024, 2031))
tam = [1.21, 1.42, 1.67, 1.95, 2.28, 2.67, 3.15]  # Trillions USD, converted from billions
sam = [18.57, 22.39, 26.98, 32.47, 38.97, 46.56, 55.43]  # Billions USD
som_initial_share = 0.05  # 5% market penetration initially
som_target_share = 0.10  # 10% market penetration target
som_initial = [x * som_initial_share for x in sam]  # Initial SOM
som_target = [x * som_target_share for x in sam]  # Target SOM by 2030

# Plotting the data
plt.figure(figsize=(12, 6))

# TAM
plt.plot(years, tam, marker='o', linestyle='-', color='b', label='Total Addressable Market (TAM)')

# SAM
plt.plot(years, sam, marker='o', linestyle='-', color='g', label='Serviceable Available Market (SAM)')

# SOM initial and target
plt.plot(years, som_initial, marker='o', linestyle='--', color='r', label='Serviceable Obtainable Market (SOM) - Initial 5%')
plt.plot(years, som_target, marker='o', linestyle='--', color='orange', label='Serviceable Obtainable Market (SOM) - Target 10%')

plt.title('Market Size Projections for Charged Motors (2024-2030)')
plt.xlabel('Year')
plt.ylabel('Market Size (in billions USD)')
plt.grid(True)
plt.xticks(years, rotation=45)
plt.legend()
plt.tight_layout()

# Display the plot
plt.show()
```
